```python
import os
import csv
import xml.etree.ElementTree as ET
import numpy as np
from shapely.geometry import Polygon

def parse_xml(file_path):
    try:
        tree = ET.parse(file_path)
        root = tree.getroot()
        ns = {'pc': 'http://schema.primaresearch.org/PAGE/gts/pagecontent/2013-07-15'}
        regions = []
        # Parsing both TextRegion and ImageRegion
        for text_region in root.findall('.//pc:TextRegion', ns):
            coords = text_region.find('.//pc:Coords', ns).attrib['points']
            points = [tuple(map(int, point.split(','))) for point in coords.split()]
            regions.append(points)

        for image_region in root.findall('.//pc:ImageRegion', ns):
            coords = image_region.find('.//pc:Coords', ns).attrib['points']
            points = [tuple(map(int, point.split(','))) for point in coords.split()]
            regions.append(points)

        return regions
    except ET.ParseError as e:
        print(f"Error parsing {file_path}: {e}")
        return []


def polygon_area(points):
    x, y = zip(*points)
    return 0.5 * np.abs(np.dot(x, np.roll(y, 1)) - np.dot(y, np.roll(x, 1)))

def intersect_area(poly1, poly2):
    poly1 = Polygon(poly1)
    poly2 = Polygon(poly2)
    if not poly1.intersects(poly2):
        return 0.0
    return poly1.intersection(poly2).area

def iou(poly1, poly2):
    inter_area = intersect_area(poly1, poly2)
    union_area = polygon_area(poly1) + polygon_area(poly2) - inter_area
    return inter_area / union_area

def evaluate(pred_folder, gt_folder, eval_files, iou_threshold=0.5):
    tp = 0
    fp = 0
    fn = 0

    for pred_file in os.listdir(pred_folder):
        if pred_file not in eval_files:
            continue

        pred_file_path = os.path.join(pred_folder, pred_file)
        gt_file_path = os.path.join(gt_folder, pred_file)
        if not os.path.exists(gt_file_path):
            print(f"Ground truth file not found for {pred_file}")
            continue

        pred_regions = parse_xml(pred_file_path)
        gt_regions = parse_xml(gt_file_path)

        if not pred_regions or not gt_regions:
            print(f"No regions found in {pred_file}")
            continue

        matched_gt = [False] * len(gt_regions)

        for pred_region in pred_regions:
            best_iou = 0
            best_gt_index = -1
            for gt_index, gt_region in enumerate(gt_regions):
                score = iou(pred_region, gt_region)
                if score > best_iou:
                    best_iou = score
                    best_gt_index = gt_index

            if best_iou >= iou_threshold:
                tp += 1
                matched_gt[best_gt_index] = True
            else:
                fp += 1

        fn += sum(1 for matched in matched_gt if not matched)

    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    accuracy = tp / (tp + fp + fn) if (tp + fp + fn) > 0 else 0

    return precision, recall, f1_score, accuracy

def load_eval_files(csv_path):
    eval_files = set()
    with open(csv_path, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            if row['split_index'] == '0':
                eval_files.add(row['hashed_prefix'] + '.xml')
    return eval_files

# Define the paths to your folders and CSV file
pred_folder = '/content/BE-Arabic-9K/output_data/xml'
gt_folder = '/content/BE-Arabic-9K/results/xml'
csv_path = '/content/BE-Arabic-9K/data_info.csv'

# Load evaluation filenames from CSV
eval_files = load_eval_files(csv_path)
print(f"Evaluation files: {eval_files}")

# Evaluate
precision, recall, f1_score, accuracy = evaluate(pred_folder, gt_folder, eval_files)
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1_score}')
print(f'Accuracy: {accuracy}')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def shi_tomashi(image):
    """
    Use Shi-Tomashi algorithm to detect corners

    Args:
        image: np.array

    Returns:
        corners: list

    """
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    corners = cv2.goodFeaturesToTrack(gray, 4, 0.01, 100)
    corners = np.int0(corners)
    corners = sorted(np.concatenate(corners).tolist())
    print('\nThe corner points are...\n')

    im = image.copy()
    for index, c in enumerate(corners):
        x, y = c
        cv2.circle(im, (x, y), 3, 255, -1)
        character = chr(65 + index)
        print(character, ':', c)
        cv2.putText(im, character, tuple(c), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2, cv2.LINE_AA)

    plt.imshow(im)
    plt.title('Corner Detection: Shi-Tomashi')
    plt.show()
    return corners


def get_destination_points(corners):
    """
    -Get destination points from corners of warped images
    -Approximating height and width of the rectangle: we take maximum of the 2 widths and 2 heights

    Args:
        corners: list

    Returns:
        destination_corners: list
        height: int
        width: int

    """

    w1 = np.sqrt((corners[0][0] - corners[1][0]) ** 2 + (corners[0][1] - corners[1][1]) ** 2)
    w2 = np.sqrt((corners[2][0] - corners[3][0]) ** 2 + (corners[2][1] - corners[3][1]) ** 2)
    w = max(int(w1), int(w2))

    h1 = np.sqrt((corners[0][0] - corners[2][0]) ** 2 + (corners[0][1] - corners[2][1]) ** 2)
    h2 = np.sqrt((corners[1][0] - corners[3][0]) ** 2 + (corners[1][1] - corners[3][1]) ** 2)
    h = max(int(h1), int(h2))

    destination_corners = np.float32([(0, 0), (w - 1, 0), (0, h - 1), (w - 1, h - 1)])

    print('\nThe destination points are: \n')
    for index, c in enumerate(destination_corners):
        character = chr(65 + index) + "'"
        print(character, ':', c)

    print('\nThe approximated height and width of the original image is: \n', (h, w))
    return destination_corners, h, w


def unwarp(img, src, dst):
    """

    Args:
        img: np.array
        src: list
        dst: list

    Returns:
        un_warped: np.array

    """
    h, w = img.shape[:2]
    H, _ = cv2.findHomography(src, dst, method=cv2.RANSAC, ransacReprojThreshold=3.0)
    print('\nThe homography matrix is: \n', H)
    un_warped = cv2.warpPerspective(img, H, (w, h), flags=cv2.INTER_LINEAR)

    # plot

    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 8))
    # f.subplots_adjust(hspace=.2, wspace=.05)
    ax1.imshow(img)
    ax1.set_title('Original Image')

    x = [src[0][0], src[2][0], src[3][0], src[1][0], src[0][0]]
    y = [src[0][1], src[2][1], src[3][1], src[1][1], src[0][1]]

    ax2.imshow(img)
    ax2.plot(x, y, color='yellow', linewidth=3)
    ax2.set_ylim([h, 0])
    ax2.set_xlim([0, w])
    ax2.set_title('Target Area')

    plt.show()
    return un_warped


def example_one():
    """
    Skew correction using homography and corner detection using Shi-Tomashi corner detector

    Returns: None

    """
    image = cv2.imread('/content/BE-Arabic-9K/test_data/images (1).jpg')
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title('Original Image')
    plt.show()

    corners = shi_tomashi(image)
    destination, h, w = get_destination_points(corners)
    un_warped = unwarp(image, np.float32(corners), destination)
    cropped = un_warped[0:h, 0:w]

    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5), facecolor='w', edgecolor='k')
    # f.subplots_adjust(hspace=.2, wspace=.05)

    ax1.imshow(un_warped)
    ax2.imshow(cropped)

    plt.show()


def apply_filter(image):
    """
    Define a 5X5 kernel and apply the filter to gray scale image
    Args:
        image: np.array

    Returns:
        filtered: np.array

    """
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    kernel = np.ones((5, 5), np.float32) / 15
    filtered = cv2.filter2D(gray, -1, kernel)
    plt.imshow(cv2.cvtColor(filtered, cv2.COLOR_BGR2RGB))
    plt.title('Filtered Image')
    plt.show()
    return filtered

def apply_threshold(filtered):
    """
    Apply OTSU threshold

    Args:
        filtered: np.array

    Returns:
        thresh: np.array

    """
    ret, thresh = cv2.threshold(filtered, 250, 255, cv2.THRESH_OTSU)
    plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_BGR2RGB))
    plt.title('After applying OTSU threshold')
    plt.show()
    return thresh

def detect_contour(img, image_shape):
    """

    Args:
        img: np.array()
        image_shape: tuple

    Returns:
        canvas: np.array()
        cnt: list

    """
    canvas = np.zeros(image_shape, np.uint8)
    contours, hierarchy = cv2.findContours(img, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
    cnt = sorted(contours, key=cv2.contourArea, reverse=True)[0]
    cv2.drawContours(canvas, cnt, -1, (0, 255, 255), 3)
    plt.title('Largest Contour')
    plt.imshow(canvas)
    plt.show()

    return canvas, cnt

def detect_corners_from_contour(canvas, cnt):
    """
    Detecting corner points form contours using cv2.approxPolyDP()
    Args:
        canvas: np.array()
        cnt: list

    Returns:
        approx_corners: list

    """
    epsilon = 0.02 * cv2.arcLength(cnt, True)
    approx_corners = cv2.approxPolyDP(cnt, epsilon, True)
    cv2.drawContours(canvas, approx_corners, -1, (255, 255, 0), 10)
    approx_corners = sorted(np.concatenate(approx_corners).tolist())
    print('\nThe corner points are ...\n')
    for index, c in enumerate(approx_corners):
        character = chr(65 + index)
        print(character, ':', c)
        cv2.putText(canvas, character, tuple(c), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)

    # Rearranging the order of the corner points
    approx_corners = [approx_corners[i] for i in [0, 2, 1, 3]]

    plt.imshow(canvas)
    plt.title('Corner Points: Douglas-Peucker')
    plt.show()
    return approx_corners

def preprocessing(image):
    """
    Skew correction using homography and corner detection using contour points
    Returns: None

    """

    plt.imshow(image)
    plt.title('Original Image')
    plt.show()

    filtered_image = apply_filter(image)
    threshold_image = apply_threshold(filtered_image)

    cnv, largest_contour = detect_contour(threshold_image, image.shape)
    corners = detect_corners_from_contour(cnv, largest_contour)

    destination_points, h, w = get_destination_points(corners)
    un_warped = unwarp(image, np.float32(corners), destination_points)

    cropped = un_warped[0:h, 0:w]
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 8))
    f.subplots_adjust(hspace=.2, wspace=.05)
    ax1.imshow(un_warped)
    ax2.imshow(cropped)

    plt.show()
    return cropped

if __name__ == '__main__':
    # example_one()
    image_path = '/content/BE-Arabic-9K/output_data/jpg/055f78940c07630352676197ab7c3ed8d5bb204d406e5c6ff101f870bc9b7dd7.jpg'
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    preprocessing(image)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def deskew_image(image):
    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply binary thresholding
    _, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # Find contours
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Get the largest contour
    largest_contour = max(contours, key=cv2.contourArea)

    # Get the minimum area rectangle for the largest contour
    rect = cv2.minAreaRect(largest_contour)
    box = cv2.boxPoints(rect)
    box = np.int0(box)

    # Calculate the angle of rotation
    angle = rect[2]
    if angle < -45:
        angle = 90 + angle

    # Get the image center
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)

    # Calculate the rotation matrix
    M = cv2.getRotationMatrix2D(center, angle, 1.0)

    # Perform the rotation
    rotated = cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_CUBIC, borderMode=cv2.BORDER_REPLICATE)

    return rotated, angle

# Load the image
image_path = '/content/BE-Arabic-9K/test_data/images.jpg'
image = cv2.imread(image_path)

# Deskew the image
deskewed_image, skew_angle = deskew_image(image)

# Display the original and deskewed images
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title('Deskewed Image')
plt.imshow(cv2.cvtColor(deskewed_image, cv2.COLOR_BGR2RGB))

plt.show()

print(f'Skew angle: {skew_angle} degrees')
```

```python
import os
import cv2
import csv
import torch
import datetime
from PIL import Image
import torchvision.transforms as transforms
import torchvision
from xml.etree.ElementTree import Element, SubElement, ElementTree
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor

class Args:
    def __init__(self):
        self.dataset_dir = "/content/BE-Arabic-9K/test_data"
        self.data_info_path = "data_info.csv"
        self.evaluation_set = 0
        self.batch_size = 4
        self.region_acceptance_score = 0.8
        self.i_r_threshold = 0.2
        self.model_path = "./testing_model"
        self.visualization_output_directory = "/content/BE-Arabic-9K/test_output"
        self.xml_output_directory = "results/xml"
        # self.image_path = "/content/BE-Arabic-9K/output_data/jpg/00037f39cf870a1f49129f9c82d935665d352ffd25ea3296208f6f7b16fd654f.jpg"  # Replace with actual image path

args = Args()

class Point:
    def __init__(self, setX, setY):
        self.x = setX
        self.y = setY

    def printPoint(self):
        print("x: " + str(self.x))
        print("y: " + str(self.y))

class BoundingBox:
    def __init__(self, set_logical_class, set_x_left, set_x_right, set_y_top, set_y_bottom):
        self.logical_class = set_logical_class
        self.x_left = set_x_left
        self.x_right = set_x_right
        self.y_top = set_y_top
        self.y_bottom = set_y_bottom

    def printBox(self):
        print("logical_class: " + str(self.logical_class))
        print("x_left: " + str(self.x_left))
        print("x_right: " + str(self.x_right))
        print("y_top: " + str(self.y_top))
        print("y_bottom: " + str(self.y_bottom))

def create_testing_input(img_path, device):

    transform = transforms.Compose([transforms.ToTensor()])
    img = Image.open(img_path).convert("RGB")
    #
    # Creating the kernel(2d convolution matrix)
    # kernel1 = np.ones((5, 5), np.float32)/30
    img = np.array(img)
    # Applying the filter2D() function
    # img = cv2.filter2D(src=img, ddepth=-1, kernel=kernel1)

    # img = preprocessing(img)
    image_tensor = transform(img)
    image_tensor_gpu = image_tensor.to(device)
    return image_tensor_gpu

def load_model(model_path):
    model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=True)
    num_classes = 3
    in_features = model.roi_heads.box_predictor.cls_score.in_features
    model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes)
    new_aspect_ratios = (0.5, 1.0, 2.0)
    new_aspect_ratios = (new_aspect_ratios,) * len(model.rpn.anchor_generator.sizes)
    model.rpn.anchor_generator.aspect_ratios = new_aspect_ratios
    model.load_state_dict(torch.load(model_path))
    print("FFRA model load complete.")
    return model

def draw_boxes_in_image(image, boxes_text, boxes_picture, out_path):
    def draw_boxes(image, boxes, color):
        for box in boxes:
            xmin = int(box.x_left)
            xmax = int(box.x_right)
            ymin = int(box.y_top)
            ymax = int(box.y_bottom)
            start_point = (xmin, ymin)
            end_point = (xmax, ymax)
            thickness = 9
            image = cv2.rectangle(image, start_point, end_point, color, thickness)
        return image

    text_color = (255, 0, 0)
    picture_color = (0, 255, 0)

    image = draw_boxes(image, boxes_text, text_color)
    image = draw_boxes(image, boxes_picture, picture_color)
    cv2.imwrite(out_path, image)
    print(f"Image saved to {out_path}")

    return image

def merging_intersected_boxes(boxes, i_r_threshold):
    i = 0
    j = 0
    length_of_boxes_list = len(boxes)
    while i < length_of_boxes_list:
        if i != j:
            box_A = boxes[i]
            box_B = boxes[j]
            intersect_ratio = ratio_of_intersection_to_small_box(box_A, box_B)
            if intersect_ratio > i_r_threshold:
                x_left = min(int(box_A.x_left), int(box_B.x_left))
                x_right = max(int(box_A.x_right), int(box_B.x_right))
                y_top = min(int(box_A.y_top), int(box_B.y_top))
                y_bottom = max(int(box_A.y_bottom), int(box_B.y_bottom))
                logical_class = int(box_A.logical_class)
                boxes[i] = BoundingBox(logical_class, x_left, x_right, y_top, y_bottom)
                del boxes[j]
                length_of_boxes_list = len(boxes)
                i = 0
                j = 0
            else:
                j += 1
                if j >= length_of_boxes_list:
                    j = 0
                    i += 1
        else:
            j += 1
            if j >= length_of_boxes_list:
                j = 0
                i += 1

    return boxes

def intersected_area(box_A, box_B):
    dx = min(box_A.x_right, box_B.x_right) - max(box_A.x_left, box_B.x_left)
    dy = min(box_A.y_bottom, box_B.y_bottom) - max(box_A.y_top, box_B.y_top)
    if (dx >= 0) and (dy >= 0):
        return dx * dy
    else:
        return 0

def ratio_of_intersection_to_small_box(box_A, box_B):
    area_intersected = intersected_area(box_A, box_B)
    area_A = (box_A.x_right - box_A.x_left) * (box_A.y_bottom - box_A.y_top)
    area_B = (box_B.x_right - box_B.x_left) * (box_B.y_bottom - box_B.y_top)

    if area_A > area_B:
        ratio = area_intersected / area_B
    else:
        ratio = area_intersected / area_A

    return ratio

def write_xml(image, image_name, boxes_text, boxes_picture, out_file_name):
    time = datetime.datetime.today()
    PcGts = Element("PcGts")
    PcGts.set('xmlns', 'http://schema.primaresearch.org/PAGE/gts/pagecontent/2013-07-15')
    PcGts.set('xmlns:xsi', 'http://www.w3.org/2001/XMLSchema-instance')
    PcGts.set('xsi:schemaLocation', 'http://schema.primaresearch.org/PAGE/gts/pagecontent/2013-07-15 http://schema.primaresearch.org/PAGE/gts/pagecontent/2013-07-15/pagecontent.xsd')
    Metadata = SubElement(PcGts, "Metadata")
    Creator = SubElement(Metadata, "Creator")
    Created = SubElement(Metadata, "Created").text = str(time.year) + "-" + str(time.month) + "-" + str(time.day) + "T" + str(time.hour) + ":" + str(time.minute) + ":" + str(time.second)
    LastChange = SubElement(Metadata, "LastChange").text = str(time.year) + "-" + str(time.month) + "-" + str(time.day) + "T" + str(time.hour) + ":" + str(time.minute) + ":" + str(time.second)
    Page = SubElement(PcGts, "Page")
    Page.set('imageFilename', image_name)
    height = image.shape[0]
    width = image.shape[1]
    Page.set('imageWidth', str(width))
    Page.set('imageHeight', str(height))
    index_count = 0
    for box in boxes_text:
        index_count += 1
        xl = box.x_left
        if xl <= 0:
            xl += 1
        xr = box.x_right
        yt = box.y_top
        if yt <= 0:
            yt += 1
        yb = box.y_bottom
        TextRegion = SubElement(Page, 'TextRegion')
        TextRegion.set('id', "r" + str(index_count))
        Coords = SubElement(TextRegion, 'Coords')
        Coords.set('points', str(xl) + "," + str(yt) + " " + str(xl) + "," + str(yb) + " " + str(xr) + "," + str(yb) + " " + str(xr) + "," + str(yt))

    for box in boxes_picture:
        index_count += 1
        xl = box.x_left
        if xl <= 0:
            xl += 1
        xr = box.x_right
        yt = box.y_top
        if yt <= 0:
            yt += 1
        yb = box.y_bottom
        ImageRegion = SubElement(Page, 'ImageRegion')
        ImageRegion.set('id', "r" + str(index_count))
        Coords = SubElement(ImageRegion, 'Coords')
        Coords.set('points', str(xl) + "," + str(yt) + " " + str(xl) + "," + str(yb) + " " + str(xr) + "," + str(yb) + " " + str(xr) + "," + str(yt))

    file = open(out_file_name, 'wb')
    # Convert the string to bytes before writing
    file.write(b'<?xml version="1.0" encoding="UTF-8"?>' + b"\n")
    ElementTree(PcGts).write(file, encoding='utf-8')  # Specify encoding here
    print(f"XML saved to {out_file_name}")
    file.close()

def get_predictions(output, region_acceptance_score):
    boxes_text_list = []
    boxes_picture_list = []
    if len(output[0]['labels']) != 0:
        for i in range(len(output[0]['labels'])):
            label = int(output[0]['labels'][i])
            score = float(output[0]['scores'][i])
            x_left = float(output[0]['boxes'][i][0])
            y_top = float(output[0]['boxes'][i][1])
            x_right = float(output[0]['boxes'][i][2])
            y_bottom = float(output[0]['boxes'][i][3])
            if score >= region_acceptance_score:
                if label == 1:
                    boxes_text_list.append(BoundingBox(label, x_left, x_right, y_top, y_bottom))
                elif label == 2:
                    boxes_picture_list.append(BoundingBox(label, x_left, x_right, y_top, y_bottom))

    return boxes_text_list, boxes_picture_list

# Ensure the output directories exist
os.makedirs(args.visualization_output_directory, exist_ok=True)
os.makedirs(args.xml_output_directory, exist_ok=True)

def main(args):
    device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
    model = load_model(args.model_path)
    model.eval()
    model.to(device)

    print(f"Checking files in {args.dataset_dir}...")
    files = os.listdir(args.dataset_dir)
    print(f"Files found: {files}")


    images_list = [img for img in sorted(os.listdir(args.dataset_dir)) if img.endswith(".jpg")]
    print(f"Found {len(images_list)} images.")
    for image_name in images_list:
        print(f"Processing {image_name}...")
        image_tensor_gpu = create_testing_input(os.path.join(args.dataset_dir, image_name), device)
        output = model([image_tensor_gpu])
        boxes_text_list, boxes_picture_list = get_predictions(output, args.region_acceptance_score)
        merged_boxes_text_list = merging_intersected_boxes(boxes_text_list, args.i_r_threshold)
        merged_boxes_picture_list = merging_intersected_boxes(boxes_picture_list, args.i_r_threshold)
        image = cv2.imread(os.path.join(args.dataset_dir, image_name))
        draw_boxes_in_image(image, merged_boxes_text_list, merged_boxes_picture_list, os.path.join(args.visualization_output_directory, image_name))
        write_xml(image, image_name, merged_boxes_text_list, merged_boxes_picture_list, os.path.join(args.xml_output_directory, image_name.replace(".jpg", ".xml")))

# Call the main function
main(args)
```

```python
import os
import cv2
import csv
import torch
import datetime
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
import torchvision
from xml.etree.ElementTree import Element, SubElement, ElementTree
from xml.dom import minidom
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor


class Args:
    def __init__(self):
        self.dataset_dir = "/content/BE-Arabic-9K/skewed_data"
        self.data_info_path = "data_info.csv"
        self.evaluation_set = 0
        self.batch_size = 4
        self.region_acceptance_score = 0.8
        self.i_r_threshold = 0.2
        self.model_path = "./testing_model"
        self.visualization_output_directory = "/content/BE-Arabic-9K/test_output"
        self.xml_output_directory = "results/xml"


args = Args()


class Point:
    def __init__(self, setX, setY):
        self.x = setX
        self.y = setY

    def printPoint(self):
        print("x: " + str(self.x))
        print("y: " + str(self.y))


class BoundingBox:
    def __init__(self, set_logical_class, set_x_left, set_x_right, set_y_top, set_y_bottom):
        self.logical_class = set_logical_class
        self.x_left = set_x_left
        self.x_right = set_x_right
        self.y_top = set_y_top
        self.y_bottom = set_y_bottom

    def printBox(self):
        print("logical_class: " + str(self.logical_class))
        print("x_left: " + str(self.x_left))
        print("x_right: " + str(self.x_right))
        print("y_top: " + str(self.y_top))
        print("y_bottom: " + str(self.y_bottom))


def deskew_image(image):
    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply binary thresholding
    _, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # Find contours
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Get the largest contour
    largest_contour = max(contours, key=cv2.contourArea)

    # Get the minimum area rectangle for the largest contour
    rect = cv2.minAreaRect(largest_contour)
    box = cv2.boxPoints(rect)
    box = np.int0(box)

    # Calculate the angle of rotation
    angle = rect[2]
    if angle < -45:
        angle = 90 + angle

    # Get the image center
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)

    # Calculate the rotation matrix
    M = cv2.getRotationMatrix2D(center, angle, 1.0)

    # Perform the rotation
    rotated = cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_CUBIC, borderMode=cv2.BORDER_REPLICATE)

    return rotated, M


def create_testing_input(img_path, device):
    # Load the image
    img = Image.open(img_path).convert("RGB")
    img = np.array(img)

    # Deskew the image
    deskewed_img, rotation_matrix = deskew_image(img)

    # Show the original and deskewed images (for visualization, if needed)
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(img)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title(f"Deskewed Image")
    plt.imshow(deskewed_img)
    plt.axis('off')

    plt.show()

    # Convert the deskewed image back to PIL format
    deskewed_img = Image.fromarray(deskewed_img)

    # Save the deskewed image to a new file, keeping the original skewed image intact
    # Create a new filename by appending '_deskewed' to the original image name
    file_dir, file_name = os.path.split(img_path)
    file_name_without_extension, file_extension = os.path.splitext(file_name)
    new_file_name = f"{file_name_without_extension}_deskewed{file_extension}"
    new_img_path = os.path.join(file_dir, new_file_name)

    deskewed_img.save(new_img_path)  # Save the deskewed image with the new filename

    print(f"Deskewed image saved as: {new_img_path}")

    # Transform the deskewed image for model input
    transform = transforms.Compose([transforms.ToTensor()])
    image_tensor = transform(deskewed_img)
    image_tensor_gpu = image_tensor.to(device)

    return image_tensor_gpu, rotation_matrix, new_img_path  # Also return the path to the deskewed image




def load_model(model_path):
    model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=True)
    num_classes = 3  # Adjust this depending on the number of classes in your dataset
    in_features = model.roi_heads.box_predictor.cls_score.in_features
    model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes)
    model.load_state_dict(torch.load(model_path))
    model.eval()
    print("Model loaded successfully.")
    return model


def draw_boxes_in_image(image, boxes_text, boxes_picture, out_path):
    def draw_boxes(image, boxes, color):
        for box in boxes:
            xmin = int(box.x_left)
            xmax = int(box.x_right)
            ymin = int(box.y_top)
            ymax = int(box.y_bottom)
            start_point = (xmin, ymin)
            end_point = (xmax, ymax)
            thickness = 9
            image = cv2.rectangle(image, start_point, end_point, color, thickness)
        return image

    text_color = (255, 0, 0)
    picture_color = (0, 255, 0)

    image = draw_boxes(image, boxes_text, text_color)
    image = draw_boxes(image, boxes_picture, picture_color)
    cv2.imwrite(out_path, image)
    print(f"Image saved to {out_path}")

    return image


def merging_intersected_boxes(boxes, i_r_threshold):
    i = 0
    j = 0
    length_of_boxes_list = len(boxes)
    while i < length_of_boxes_list:
        if i != j:
            box_A = boxes[i]
            box_B = boxes[j]
            intersect_ratio = ratio_of_intersection_to_small_box(box_A, box_B)
            if intersect_ratio > i_r_threshold:
                x_left = min(int(box_A.x_left), int(box_B.x_left))
                x_right = max(int(box_A.x_right), int(box_B.x_right))
                y_top = min(int(box_A.y_top), int(box_B.y_top))
                y_bottom = max(int(box_A.y_bottom), int(box_B.y_bottom))
                logical_class = int(box_A.logical_class)
                boxes[i] = BoundingBox(logical_class, x_left, x_right, y_top, y_bottom)
                del boxes[j]
                length_of_boxes_list = len(boxes)
                i = 0
                j = 0
            else:
                j += 1
                if j >= length_of_boxes_list:
                    j = 0
                    i += 1
        else:
            j += 1
            if j >= length_of_boxes_list:
                j = 0
                i += 1

    return boxes


def intersected_area(box_A, box_B):
    dx = min(box_A.x_right, box_B.x_right) - max(box_A.x_left, box_B.x_left)
    dy = min(box_A.y_bottom, box_B.y_bottom) - max(box_A.y_top, box_B.y_top)
    if (dx >= 0) and (dy >= 0):
        return dx * dy
    else:
        return 0


def ratio_of_intersection_to_small_box(box_A, box_B):
    area_intersected = intersected_area(box_A, box_B)
    area_A = (box_A.x_right - box_A.x_left) * (box_A.y_bottom - box_A.y_top)
    area_B = (box_B.x_right - box_B.x_left) * (box_B.y_bottom - box_B.y_top)

    if area_A > area_B:
        ratio = area_intersected / area_B
    else:
        ratio = area_intersected / area_A

    return ratio


def write_xml(image, image_name, boxes_text, boxes_picture, out_file_name):
    time = datetime.datetime.today()
    PcGts = Element("PcGts")
    PcGts.set('xmlns', 'http://schema.primaresearch.org/PAGE/gts/pagecontent/2013-07-15')
    PcGts.set('xmlns:xsi', 'http://www.w3.org/2001/XMLSchema-instance')
    PcGts.set('xsi:schemaLocation', 'http://schema.primaresearch.org/PAGE/gts/pagecontent/2013-07-15 https://schema.primaresearch.org/PAGE/gts/pagecontent/2013-07-15/pagecontent.xsd')
    Metadata = SubElement(PcGts, "Metadata")
    Creator = SubElement(Metadata, "Creator")
    Creator.text = "Anonymous"
    Created = SubElement(Metadata, "Created")
    Created.text = str(time.year) + '-' + str(time.month) + '-' + str(time.day)
    LastChange = SubElement(Metadata, "LastChange")
    LastChange.text = str(time.year) + '-' + str(time.month) + '-' + str(time.day)
    Page = SubElement(PcGts, "Page")
    Page.set("imageFilename", image_name)
    Page.set("imageWidth", str(image.shape[1]))
    Page.set("imageHeight", str(image.shape[0]))
    TextRegionId = 0
    ImageRegionId = 0

    for box in boxes_text:
        TextRegionId += 1
        TextRegion = SubElement(Page, "TextRegion")
        TextRegion.set("id", "r" + str(TextRegionId))
        Coords = SubElement(TextRegion, "Coords")
        Coords.set("points", str(box.x_left) + ',' + str(box.y_top) + ' ' +
                   str(box.x_right) + ',' + str(box.y_top) + ' ' +
                   str(box.x_right) + ',' + str(box.y_bottom) + ' ' +
                   str(box.x_left) + ',' + str(box.y_bottom))

    for box in boxes_picture:
        ImageRegionId += 1
        ImageRegion = SubElement(Page, "ImageRegion")
        ImageRegion.set("id", "r" + str(ImageRegionId))
        Coords = SubElement(ImageRegion, "Coords")
        Coords.set("points", str(box.x_left) + ',' + str(box.y_top) + ' ' +
                   str(box.x_right) + ',' + str(box.y_top) + ' ' +
                   str(box.x_right) + ',' + str(box.y_bottom) + ' ' +
                   str(box.x_left) + ',' + str(box.y_bottom))

    tree = ElementTree(PcGts)
    tree.write(out_file_name)
    print(f"XML saved to {out_file_name}")



def main():
    # Set device for model inference (GPU if available, otherwise CPU)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # Load the model
    model = load_model(args.model_path)
    model.to(device)

    # Process a specific image from your dataset
    image_path = os.path.join(args.dataset_dir, "skewed_text.jpg")  # Replace with your image path
    image_tensor_gpu, rotation_matrix, deskewed_image_path = create_testing_input(image_path, device)

    # Perform inference with the model
    with torch.no_grad():
        prediction = model([image_tensor_gpu])

    # Extract the bounding boxes and scores
    boxes = prediction[0]['boxes'].cpu().numpy()
    scores = prediction[0]['scores'].cpu().numpy()

    # Filter out boxes with low confidence scores
    filtered_boxes = []
    for i in range(len(scores)):
        if scores[i] > args.region_acceptance_score:
            x_left, y_top, x_right, y_bottom = boxes[i]
            filtered_boxes.append(BoundingBox(0, x_left, x_right, y_top, y_bottom))  # logical_class is set to 0 for simplicity

    # Merge intersecting boxes
    merged_boxes = merging_intersected_boxes(filtered_boxes, args.i_r_threshold)

    # Draw the boxes on the deskewed image
    output_image = draw_boxes_in_image(np.array(Image.open(deskewed_image_path)), merged_boxes, [], args.visualization_output_directory + "/deskewed_text.jpg")

    # Save the results to XML
    write_xml(np.array(Image.open(deskewed_image_path)), "sample_image.png", merged_boxes, [], args.xml_output_directory + "/output.xml")

    print("Processing complete.")



if __name__ == "__main__":
    main()
```
